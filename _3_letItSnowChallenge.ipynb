## **Let It Snow Challenge**

In this challenge we will create a randomised Christmas cards using Python Turtle. Our card will include 20 snowflakes of random sizes, colours and number of branches, randomly positioned on the canvas.

We will also add some Christmas Greetings to our card.

The aim of this challenge is to learn how to define your own functions in Python and to add your own parameters to your functions.

Before completing this challenge, you should first learn how to draw a snowflake using Python Turtle by completing our snowflake challenge. (see: https://github.com/AzizKABA/pythonCodeChallenges/blob/main/_2_snowflakeChallenge.ipynb)

# **My Solution**

```python
# RANDOM SNOWFLAKE POSTCARD

# Import libraries:

import random
import turtle

# Define the constant values

# NOTE: Python does not support constants as a type

NUMBER_OF_SNOWFLAKES = 20 # how many snowflakes does postcard have?


# Drawing Environments Settings
    
penObj = turtle.Turtle() # create a pen object from turtle class to draw snowflakes

penObj.speed("fastest") # set the drawing speed to the fastest

penObj.hideturtle() # hide turtle image
    
turtle.colormode(1) # rgb values are inside (0,1)

# draw the line back and ahead of the current position
def fwdBackfwd(SMALL_BRANCH_SIZE):
    penObj.forward(SMALL_BRANCH_SIZE) # move pencil forward by SMALL_BRANCH_SIZE units to draw body of the small arrow line
    penObj.backward(SMALL_BRANCH_SIZE*2) # move pencil backward by 2*SMALL_BRANCH_SIZE units to draw body of the small arrow line
    penObj.forward(SMALL_BRANCH_SIZE) # move pencil forward by SMALL_BRANCH_SIZE units to draw body of the small arrow line

# the tip of the pen is pointing from origin to positive x-axis in Euclidean plane

# loop for total number of snowflakes

for snowflakeCounter in range(NUMBER_OF_SNOWFLAKES):
    
    MAIN_BRANCH_SIZE  = random.randint(20,120) # length of the main branch random integer between 20 and 120
    SMALL_BRANCH_SIZE = int(MAIN_BRANCH_SIZE/4) # length of the arrow branch depends on randomly chosen main_branch_size
    NUMBER_OF_BRANCH  = random.randint(3,10) # number of the branches is the snowflake random integer between 3 and 10
    BRANCH_ANGLE      = int(360/NUMBER_OF_BRANCH) # angle between branches, depends on branch number
    penObj.color(random.random(),random.random(),random.random()) # set a random color in (r,g,b) space
    penObj.pensize(random.randint(5,15)) # set a random pen size between 5 and 15
    penObj.penup() # move pen up to prevent drawing while going to random position
    penObj.goto(random.randint(-300,300),random.randint(-300,300)) # choose a random position for snowflakes
    penObj.pendown() # move pen back down to start drawing
    
    # loop for drawing a single snowflake
    
    for branchCounter in range(NUMBER_OF_BRANCH):
                       
        penObj.forward(MAIN_BRANCH_SIZE) # move pencil forward by MAIN_BRANCH_SIZE units to draw body of the snowflake branch
        penObj.backward(SMALL_BRANCH_SIZE) # return to the small branch's origin point
    
        penObj.right(90) # direct pen's heading 90 degrees right to draw small arrow line
        fwdBackfwd(SMALL_BRANCH_SIZE) # draw the line back and ahead of the current position
    
        penObj.right(45) # direct pen's heading 45 degrees right to draw small arrow line
        fwdBackfwd(SMALL_BRANCH_SIZE) # draw the line back and ahead of the current position
    
        penObj.right(90) # direct pen's heading 90 degrees right to draw small arrow line
        fwdBackfwd(SMALL_BRANCH_SIZE) # draw the line back and ahead of the current position
    
        penObj.right(135) # direct pen's heading 135 degrees right to match heading with main branch
        penObj.backward(MAIN_BRANCH_SIZE-SMALL_BRANCH_SIZE) # return the main origin point
        penObj.right(BRANCH_ANGLE) # jump next branch
    # end of for loop
# end of for loop
# end of program
```
