### Step 2: Solution for _Using a For Loop to complete the snowflake_

```python
# First import turtle library to draw snowflakes from lines
# Additional information about library is available @ https://docs.python.org/3/library/turtle.html

import turtle

# Define the constant values

# NOTE: Python does not support constants as a type

SMALL_BRANCH_SIZE = 30 # length of the arrow branch
MAIN_BRANCH_SIZE  = 90 # length of the main branch
BRANCH_ANGLE      = 60 # there are 6 branches in a circle. 360 / 6 = 60 degrees each

# Drawing Environments Settings
    
penObj = turtle.Turtle() # create a pen object from turtle class to draw snowflakes

penObj.speed("fastest") # set the drawing speed to the fastest

penObj.hideturtle() # hide turtle image
    
penObj.color("black") # set color to black

penObj.pensize(10) # set pen size 

# the tip of the pen is pointing from origin to positive x-axis in Euclidean plane

for counter in range(6):
    penObj.forward(MAIN_BRANCH_SIZE) # move pencil forward by 90 units to draw body of the snowflake branch
    penObj.backward(SMALL_BRANCH_SIZE) # return to the small branch's origin point
    penObj.right(45) # direct pen's heading 45 degrees right to draw small arrow line
    penObj.forward(SMALL_BRANCH_SIZE) # move pencil forward by 30 units to draw body of the small right arrow line
    penObj.backward(SMALL_BRANCH_SIZE) # return to the small branch's origin point
    penObj.left(90) # direct pen's heading 45 degrees left to draw small arrow line (relative 90 degrees to current state)
    penObj.forward(SMALL_BRANCH_SIZE) # move pencil forward by 30 units to draw body of the small left arrow line
    penObj.backward(SMALL_BRANCH_SIZE) # return to the small branch's origin poin
    penObj.right(45) # direct pen's heading 45 degrees right to match heading with main branch
    penObj.backward(MAIN_BRANCH_SIZE-SMALL_BRANCH_SIZE) # return the main origin point
    penObj.right(BRANCH_ANGLE) # jump next branch
# end of for loop

```

### Solution for _Bonus Challenge: Let it Snow_

```python
# STAR TAILED SNOWFLAKE

#First import turtle library to draw snowflakes from lines
# Additional information about library is available @ https://docs.python.org/3/library/turtle.html

import turtle

# Define the constant values

# NOTE: Python does not support constants as a type

SMALL_BRANCH_SIZE = 30 # length of the arrow branch
MAIN_BRANCH_SIZE  = 120 # length of the main branch
NUMBER_OF_BRANCH  = 5 # number of the branches is the snowflake 
BRANCH_ANGLE      = int(360/NUMBER_OF_BRANCH) # there are 5 branches in a circle. 360 / 5 = 72 degrees each

# Drawing Environments Settings
    
penObj = turtle.Turtle() # create a pen object from turtle class to draw snowflakes

penObj.speed("fastest") # set the drawing speed to the fastest

penObj.hideturtle() # hide turtle image
    
penObj.color("magenta") # set color to magenta

penObj.pensize(10) # set pen size 

# draw the line back and ahead of the current position
def fwdBackfwd(SMALL_BRANCH_SIZE):
    penObj.forward(SMALL_BRANCH_SIZE) # move pencil forward by SMALL_BRANCH_SIZE units to draw body of the small arrow line
    penObj.backward(SMALL_BRANCH_SIZE*2) # move pencil backward by 2*SMALL_BRANCH_SIZE units to draw body of the small arrow line
    penObj.forward(SMALL_BRANCH_SIZE) # move pencil forward by SMALL_BRANCH_SIZE units to draw body of the small arrow line

# the tip of the pen is pointing from origin to positive x-axis in Euclidean plane

for counter in range(NUMBER_OF_BRANCH):
    
    penObj.forward(MAIN_BRANCH_SIZE) # move pencil forward by MAIN_BRANCH_SIZE units to draw body of the snowflake branch
    penObj.backward(SMALL_BRANCH_SIZE) # return to the small branch's origin point
    
    penObj.right(90) # direct pen's heading 90 degrees right to draw small arrow line
    fwdBackfwd(SMALL_BRANCH_SIZE) # draw the line back and ahead of the current position
    
    penObj.right(45) # direct pen's heading 45 degrees right to draw small arrow line
    fwdBackfwd(SMALL_BRANCH_SIZE) # draw the line back and ahead of the current position
    
    penObj.right(90) # direct pen's heading 90 degrees right to draw small arrow line
    fwdBackfwd(SMALL_BRANCH_SIZE) # draw the line back and ahead of the current position
    
    penObj.right(135) # direct pen's heading 135 degrees right to match heading with main branch
    penObj.backward(MAIN_BRANCH_SIZE-SMALL_BRANCH_SIZE) # return the main origin point
    penObj.right(BRANCH_ANGLE) # jump next branch
# end of for loop
```
