# **My Solution**

Although challenge is just for 2 players,
I adapt it to become for variable number of players.

So, if you want 9 players to play with just enter the total number of players as 9 and leave the rest to the program :)

```python
# DOUBLE SIX DICE GAME

# import external libraries

import random

#--------------------------------------------------------------------#
# function to play the game.
# call the function once for each player

# arguments: playerNumber : integer -> any number of players
#          : playerName : a list that holds names of the all players
#          : playerAttempt : a list that holds number of attempts until double six occurs
# returns: since playerName and playerAttempt are global, no need to return them. 
# They are changed inside the function

def sixDiceGame(playerNumber,playerName,playerAttempt):
       
    playerName.append(input("Please enter the name of the player: "))

    dice1,dice2 = 0,0
     
    playerAttempt.append(0) 
    
    while dice1 != 6 and dice2 != 6:
        
        dice1, dice2 = random.randint(1,6),random.randint(1,6)
    
        playerAttempt[playerNumber] += 1
        
# end of function
#--------------------------------------------------------------------#

# welcome banner of the game

print("\t\t ___________________________\n"  +
      "\t\t|                           |\n" + 
      "\t\t|    DOUBLE SIX DICE GAME   |\n" +
      "\t\t|___________________________|\n"  )

# a variable for total number of players. Game is built with flexible number of players.
totalPlayerNumber = int(input("Please enter the number of total players: "))

# reserve spaces for playerName and playerAttempt lists.
# their length and elements will change dynamically according to the entered number of players
playerName = []
playerAttempt = []

# main loop of the game
# adapt the lists according to the # of players and call the game function for each player!
for i in range(totalPlayerNumber):
    
    sixDiceGame(i,playerName,playerAttempt)
    # end of for

# print the winner of the game with name and number of attempts
print("\nThe winner of the game is: " +
      str(playerName[playerAttempt.index(min(playerAttempt))]) + "\n"
      "Total number of attempts is: " +
      str(playerAttempt[playerAttempt.index(min(playerAttempt))]) + "\n")

# Note that if any player's winning attempt is equal,
# game assigns the winner with small player number
# though all the minimal elements can be found and listed individually if requested.

print("All Player Names: ")
print(playerName)
print("\n")
print("All Player Counts: ")
print(playerAttempt)
```
