```python
#######################################################################################################################
# a function to draw a circle with
# pen object. its properties: color, size, position and fill as arguments
def drawCircle(pen,color="red",size=10,x=0,y=0, fill = True):
    
    pen.hideturtle() # hide turtle shape
    pen.speed("fastest") # set line drawing speed to the fastest
    pen.color(color) # set color of the pen
    pen.penup() # pen is up to go to position and not to draw
    pen.goto(x,y) # go to given coordinates
    pen.pendown() # pen is ready to draw again
    if fill == True: # fill the circle? YES ->
        pen.begin_fill() # begin fill
        pen.circle(size) # draw the circle with given diameter
        pen.end_fill() # end fill
    else: # NO ->
        pen.circle(size) # # draw the circle with given diameter and leave empty inside
# end of function
#######################################################################################################################


#######################################################################################################################
# a function to set the background color of the screen
# arguments are screen object and color of the background
def setBackgroundColor(window,color):
    window.bgcolor(color) # set the bakground color of the window object
# end of function
#######################################################################################################################


#######################################################################################################################
# a function to write text to the postcard
# arguments are pen object. Text string. Color of the text. Font Size. X and Y position (start)
def writeCard(pen, text = "", color = "white", size=14, x=0,y=0):
    
    pen.hideturtle() # hide turtle shape
    pen.speed("fastest") # set line drawing speed to the fastest
    pen.color(color) # set color of the pen
    pen.penup() # pen is up to go to position and not to draw
    pen.goto(x,y) # go to given coordinates
    pen.pendown() # pen is ready to draw again
    pen.write(text, align = "center", font = ("Comic Sans",size,"normal")) 
    # write text line (center aligned) according to the given properties
# end of function
#######################################################################################################################


#######################################################################################################################
# a function to draw a star shape
# arguments are pen object. x and y position. color of the star. size of the star. 
def drawStar(pen, x = 0, y = 0, color = "yellow", size = 40, fill = True):
    
    pen.hideturtle() # hide turtle shape
    pen.speed("fastest") # set line drawing speed to the fastest
    pen.color(color) # set color of the pen
    pen.penup() # pen is up to go to position and not to draw
    pen.goto(x,y) # go to given coordinates
    pen.pendown() # pen is ready to draw again
    
    if fill == True: # fill the circle? YES ->
        pen.begin_fill() # begin fill
        for i in range(5): # set a loop for number of branches of the star (5)
            pen.right(144) # make a turn right 144 degrees
            pen.forward(size) # draw a line with given size
        pen.end_fill() # end fill
        
    else: # NO ->
        for i in range(5): # set a loop for number of branches of the star (5)
            pen.right(144) # make a turn right 144 degrees
            pen.forward(size) # draw a line with given size
        # end of for
    # end of if
# end of function
#######################################################################################################################
```
