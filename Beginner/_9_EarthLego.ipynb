### Using subroutines

It is good practice to use subroutines whenever possible as these can then easily be reused in other programs. Your task is to adapt your code by defining subroutines using two functions as follows:

    1. Your first subroutine should be called getVolumeOfSphere(), take one parameter called radius and apply the relevant formula to calculate and return the volume of a sphere based on the given radius.
    2. Your second subroutine should be called getVolumeOfCuboid(), take three parameters called width, length and height and apply the relevant formula to calculate and return the volume of a cuboid based on the given dimensions.

## **My Solution**

```python
# Calculation of how many lego bricks needed to build the planet Earth!


# define the constant values section

LEGO_WIDTH   = 16 # width of the lego brick [mm]
LEGO_LENGTH  = 16 # length of the lego brick [mm]
LEGO_HEIGHT  = 10 # height of the lego brick [mm]
EARTH_RADIUS = 6371 # radius of the earth [km]
KM_TO_MM     = 1000000 # convertion unit from km to mm
pi           = 3.1415926 # pi constant


# define function prototypes section

# function to calculate the volume of a sphere
# accepts radius as an argument in [km] and 
# returns the volume of the sphere

def getVolumeOfSphere(radius):
    return (4*pi*((radius*KM_TO_MM)**3))/3
# end of function


# function to calculate the volume of a cuboid
# accepts width,length and height as arguments and 
# returns the volume of the cuboid

def getVolumeOfCuboid(width,length,height):
    return width*length*height
# end of function


# calculation and user output section

# print the radius of the earth in mm
print("Radius of the Earth is: " + str(EARTH_RADIUS*KM_TO_MM) 
      + " [mm] ") 

# print the volume of the Earth in mm^3. 
print("Volume of the Earth is: " 
      + str(getVolumeOfSphere(EARTH_RADIUS)) 
      + " [mm\u00b3] ") # \u00b3 -> mm^3

# print the volume of a lego cuboid in mm^3. 
print("Volume of the lego cuboid is: " 
      + str(getVolumeOfCuboid(LEGO_WIDTH,LEGO_LENGTH,LEGO_HEIGHT)) 
      + " [mm\u00b3] ") # \u00b3 -> mm^3

# print how many legos are needed to build the Earth. 
print("\n\nTo build Earth out of legos, one needs: " 
      + str(getVolumeOfSphere(EARTH_RADIUS)/getVolumeOfCuboid(LEGO_WIDTH,LEGO_LENGTH,LEGO_HEIGHT)) 
      + " bricks!")

# end of program
```
