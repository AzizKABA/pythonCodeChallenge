## **Jack and the Beanstalk**

In the English fairy tale, Jack and the beanstalk, Jack is a young, poor boy living with his mother and a dairy cow on a farm cottage. The cow’s milk is their only source of income. When the cow stops giving milk, Jack’s mother tells him to take her to the market to be sold. On the way, Jack meets a curious man who offers magic beans in exchange for the cow, and surprisingly Jack makes the trade. When he arrives home without any money, his mother becomes angry, throws the beans out of the window, and sends Jack to bed without dinner.

During the night, the magic beans cause a gigantic beanstalk to grow outside Jack’s window. Let’s stop the fairy tale story here and focus on the magic beanstalk.

## **My Solution**

```python
"""
Jack and Beanstalk Code Challenge

Calculate the height of the beanstalk for after given hours passed
"""

beanGrowHeight = 100 # set initial value of beanHeight

# Welcome message
print("\t\t ___________________________\n"  +
      "\t\t|                           |\n" +
      "\t\t|    The Magic Beanstalk    |\n" +
      "\t\t|___________________________|\n"  )

beanGrowHours = 0 # user input variable of grow hours

#while loop to ask user for input until it is correct (hour>=1)
while (beanGrowHours < 1):
    
    beanGrowHours = int(input("\nEnter a number of hour (Hour must be equal to or greater than 1): ")) # user input
    
    if (beanGrowHours < 1): #  check the input
        
        print("(Hour must be >=1) ") # remind correct format if it is wrong
        
    else: # it is ok
        
        print("\n\nAfter 1 hour,  the beanstalk was 100 [cm] tall \n") # print height after one hour
    # end of for
# end of while

for i in range(2,beanGrowHours+1): # loop for calculating and printing bean grow height wrt given hours
    
    beanGrowHeight = beanGrowHeight*1.5 + 30 # calculate bean grow height
    
    # print the calculated value as a list in ascending order wrt hours
    print("After " + str(i) + " hours, the beanstalk was " + str(round(beanGrowHeight)) + " [cm] tall! \n")
    
# end of for

# end of program
```
